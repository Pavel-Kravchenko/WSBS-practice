```python
from Bio import SeqIO

record_dict = SeqIO.to_dict(SeqIO.parse("CA.s.fasta", "fasta"))
#  print(record_dict)
# print(record_dict.keys())

d = {}
for i in range(3999):
    d[i] = 0
for i in record_dict.keys():
    #print(i)
    #print(len(record_dict[i].seq))
    seq = record_dict[i].seq.split(",")
    c = 0
    for q in seq:
        if q == "1":
            if d[c] != 0:
                d[c] += 1
            if d[c] == 0:
                d[c] = 1
        c += 1
#print(d)

x = []
y = []
for i in d.keys():
    x.append(i)
    y.append(d[i])
#print(x)
#print(y)
d = {'col1': x, 'col2': y}
df = pd.DataFrame(data=d)
#df
```
